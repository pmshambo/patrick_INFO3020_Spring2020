```
#6
class Point:

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "The Point is ({},{})".format(self.x, self.y)

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)


p1 = Point(1, 3)
p2 = Point(3, 4)

print(p1)
print(p2)
print(p1 + p2)
```

```
#9
def main():

    def sql_connect(sql, variable='None'):
        import pymysql

        db = pymysql.connect("localhost", "root", "INFO3020", "world")
        dbcursor = db.cursor()
        if variable == 'None':
            command = dbcursor.execute(sql)
        else:
            command = dbcursor.execute(sql, variable)
        result = dbcursor.fetchall()
        dbcursor.close()
        db.close()

        return result

    def print_rows(data):
        for row in data:
            print(row)

    def get_conts():

        sql = "select continent from country"
        results = sql_connect(sql)
        continents = []
        for row in results:
            row = list(row)
            if row[0] not in continents:
                continents.append(row[0])
        return continents

    def get_pop():
        continents = get_conts()
        pops = []
        for i in range(len(continents)):
            sql = "select population from country where continent=%s"
            results = sql_connect(sql, continents[i])
            temp = []
            for row in results:
                row = list(row)
                temp.append(row[0])

            pops.append([continents[i], temp])
        return pops

    def pop_total():
        pops = get_pop()
        total = []
        for row in pops:
            count = 0
            for item in row[1]:
                count += item
            total.append([row[0], count])
        pop_sum = 0
        for row in total:
            pop_sum += row[1]
        total.append(pop_sum)
        return total

    def chart_pop():
        import pylab

        pop_tot = pop_total()
        pylab.figure("World Population", figsize=(8,8))

        labels = tuple(get_conts())
        gap = 0.1

        x = []

        for data in pop_tot[:-1]:
            x.append(data[1] / pop_tot[-1])

        explode = []
        for _ in range(len(labels)):
            explode.append(gap)

        explode = tuple(explode)

        pylab.pie(x, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
        pylab.title("World Population by Continent")

        pylab.show()

    chart_pop()

main()
```

```
#10
def main():

    def get_csv():
        import pandas as pd

        filename = "energy-production.csv"
        data = pd.read_csv(filename)

        return data

    def energy_by_year():

        data = get_csv()
        year = int(input("Please input a year (1973-2014): "))
        years = []
        for row in data['Year']:
            years.append(row)
        year_ind = years.index(year)
        cols = []
        for col in get_csv():
            cols.append(col)
        x = cols[2:]
        y = []
        for i in range(len(x)):
            y.append(data[x[i]][year_ind])
        xyt = [x, y, str(year)]

        return xyt

    def plot_data():
        import matplotlib.pyplot as plt

        data = energy_by_year()
        plt.figure("Energy Production")
        plt.bar(data[0], data[1])
        plt.xticks(rotation=-30)
        plt.title("Energy Production in "+data[2])

        plt.show()

    plot_data()
main()
```
